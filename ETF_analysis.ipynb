```python
from alpha_vantage.timeseries import TimeSeries
from datetime import datetime, timedelta
import pandas as pd
import time
```

```python
# Alpha Vantage API key
API_KEY = "AIEJJ0WV5K9LJ0HO"
```

```python
# Initialize Alpha Vantage TimeSeries client
def get_alpha_vantage_session():
    return TimeSeries(key=API_KEY, output_format="pandas")
```

```python
# Function to get the open price for a specific date
def get_open_price(ticker, date):
    ts = get_alpha_vantage_session()
    
    try:
        data, meta_data = ts.get_daily(symbol=ticker, outputsize="full")
        
        if date.strftime("%Y-%m-%d") in data.index:
            return round(data.loc[date.strftime("%Y-%m-%d"), "1. open"], 2)
        
        # If exact date isn't available, try closest available next day
        next_date = date + timedelta(days=1)
        if next_date.strftime("%Y-%m-%d") in data.index:
            return round(data.loc[next_date.strftime("%Y-%m-%d"), "1. open"], 2)
    
    except Exception as e:
        print(f"Error fetching {ticker} on {date.strftime('%Y-%m-%d')}: {e}")
    
    return None
```

```python
# Load your updated ETF CSV file
df = pd.read_csv("reference.csv")
```

```python
# Convert all date strings to datetime objects once
for i in range(1, 13):
    date_col = f"Distro_Date_{i}"
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce', format='%m/%d/%Y')
```

```python
# Process each row to fetch open prices for distributions
for index, row in df.iterrows():
    ticker = row["Ticker"]

    for i in range(1, 13):
        date_col = f"Distro_Date_{i}"
        open_col = f"Open_Price_D{i}"
        dist_date = row[date_col]

        if pd.notna(dist_date):
            open_price = get_open_price(ticker, dist_date)
            df.at[index, open_col] = open_price

            time.sleep(1)  # Rate limiting per request
```

```python
# Attempt to fetch Open_Price_Jan or fallback to earliest available price
def get_open_price_jan_or_first(ticker):
    ts = get_alpha_vantage_session()
    jan1 = datetime(datetime.today().year, 1, 1)

    price = get_open_price(ticker, jan1)
    if price is not None:
        return price
    
    # Fallback: Get the first available price
    try:
        data, meta_data = ts.get_daily(symbol=ticker, outputsize="full")
        if not data.empty:
            return round(data.iloc[0]["1. open"], 2)
    except Exception as e:
        print(f"Error fetching fallback for {ticker}: {e}")
    
    return None

df["Open_Price_Jan"] = df.apply(
    lambda row: get_open_price_jan_or_first(row["Ticker"]) if pd.isna(row.get("Open_Price_Jan")) else row["Open_Price_Jan"],
    axis=1
)
```

```python
# Save to file
df.to_csv("updated_reference.csv", index=False)
```
